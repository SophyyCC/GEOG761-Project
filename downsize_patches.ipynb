```python
"""
Efficiently resize patches from 256x256 to 128x128 using batch processing
"""
# Install required packages
!pip install dropbox pillow numpy matplotlib


from PIL import Image
import matplotlib.pyplot as plt
from pathlib import Path
import json
import random
from typing import List, Tuple, Dict
import dropbox
import io
import numpy as np
from PIL import Image
import json
import time
from concurrent.futures import ThreadPoolExecutor, as_completed

# ============================================================================
# SETUP
# ============================================================================

DROPBOX_ACCESS_TOKEN = "sl.u.AGDk6w7PXibX7P2KJq08bkqHF5vzYzM8QrMkLbEgGZ2RzCF35RM3mubLzu6CQSRu4d_5SOv1XuQ6pVDK1Eo6i7unzyIdgMu3OqejSAQNiSfu0NzxiaGR7BF9G4V9BKAQ_5IeiNi-PwPykoXMPleCLmkMxG2u-MfLvkDOfuf0mAVuyK1glV7dY6erysZHPy_PcL_ewAreogBT9LnIp39I7P_YwKwU1QlM00iHoQqhKjiyDIhjdBYLdKmAMO3K2gJVFlg8DFrU00iEUxyyHnm6j4mQpz6iU3NVBO4pfoBWSqq0VHdHj8a1n3sJiyoQUUCGLpmMOcECqb25JKrC9szSEFO_VoHiwv-OGSz7dcltCX_8NxHmo_DYBmy7Nm67RMvueSiKdFF1DzdYtPSr_2hQwt-W_eXXp-QsZw2YkFENeoFZm0xzeup3WoxLKNq1zEjHGKRiaw4gIb4WRfGjS_wh8lJPB9GnxuMMlFKW3i-dPXmq3WsU7b69CgdvLoF4KhmcT1Vn7xwSfZlJJ5Qvpn1JFP70JqysJnd1n8TQMtJeAQ9YL9k8y0zo0CzmeJJfMvlsnwfZXK-JxzReZZXChvWu-X7hvArAZwXWfcYhu2PDEVd6t8Ud6WG14gEcIGtPnBwS1ybUCDVeIssgsXgoCVQ6sqRH8cnkGV63iK-egS4kdJTiij0U4A4KzefKKoj4poc-B6Q8R8CvMJ515Wt2xUOuRXSdoN-Sautp98x7ofGRUIXjXi3amYe9DXVk2WoTencKaBkpiSnOAI2xhf65arPDGe2swzLrHbxki5_4crjN3dPmwdQesJVqwfYUdOBhof4ahQ-RYuDfE2Owl3jWoVNZCAq35jMlGIgU9J7wd1h6LP20p_otcFce9X1qu4ZBKtpUxSmR0RoFicd_WG9VaW3b1-XJO456ZjZPm4JS_AoQAT1VIjfK63irasbK_cfuiydpHQ_DOOPzMUEVCCGW4FwF5Wt-TGfoBq75Rt9nsJkSzaeF-_6Ill2J3XLOPsUjlMIlEGeAnKFIuCCT-kpZ5gaY3kS2cx1e-TQneGh6CZ8EnWH6KCUmcq8eDS1bN8G6e0LfqO-xO50m8majCzihDyc2zVDFqyToGf1Dg-kfYPuRD6KOge0ix1K4CntcajjDJad4rPfXwOPlgd3zzkqcVX3giq3Aq-PPezLfhe8fz9_HDS-5VUjhpCCVVcArKBWlc5WMBZfo7tG4uavVkCOXje36K8iCuIxt5XEskJ5YReAdIiKAnTMt45V41AMVc6o-tSyMgpAuY5yEJTuGd7Ikv6he4Buu-TgBFZSx2NL71DZDGkyxfLWiN8pUilUAvC08VcrpufGlkj6UxHqmIWPkZUeogR2UlZrT5p39S_CdqKu15EfMdiHUu3H2p-WUAJio1gN3EbeU7_hNOybKAXOEAZzCZ9TJ"  # Replace with your token
dbx = dropbox.Dropbox(DROPBOX_ACCESS_TOKEN)

# ============================================================================
# EFFICIENT BATCH RESIZE
# ============================================================================

def resize_single_patch(dbx, source_path, dest_path, new_size=(128, 128)):
    """
    Resize a single patch (for parallel processing)
    """
    try:
        # Download
        _, response = dbx.files_download(source_path)
        img = Image.open(io.BytesIO(response.content))

        # Resize
        resized_img = img.resize(new_size, Image.Resampling.LANCZOS)

        # Convert to bytes
        buffer = io.BytesIO()
        resized_img.save(buffer, format='PNG', optimize=True)
        buffer.seek(0)

        # Upload
        dbx.files_upload(buffer.read(), dest_path,
                       mode=dropbox.files.WriteMode.overwrite)

        return True
    except Exception as e:
        print(f"  ERROR with {source_path}: {e}")
        return False

def resize_tile_batch(dbx, tile_id, risk, num_patches,
                     source_base_path, dest_base_path,
                     new_size=(128, 128), max_workers=10):
    """
    Resize all patches from one tile using parallel processing
    """
    print(f"\n  Processing {num_patches} patches using {max_workers} parallel workers...")

    start_time = time.time()
    success_count = 0

    # Create list of all patch paths
    tasks = []
    for patch_idx in range(num_patches):
        source_path = f"{source_base_path}/{risk}/{tile_id}/patch_{patch_idx:04d}.png"
        dest_path = f"{dest_base_path}/{risk}/{tile_id}/patch_{patch_idx:04d}.png"
        tasks.append((source_path, dest_path))

    # Process in parallel
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        # Submit all tasks
        future_to_idx = {
            executor.submit(resize_single_patch, dbx, src, dst, new_size): idx
            for idx, (src, dst) in enumerate(tasks)
        }

        # Process completed tasks
        for future in as_completed(future_to_idx):
            idx = future_to_idx[future]
            try:
                if future.result():
                    success_count += 1

                # Progress update every 50 patches
                if (success_count) % 50 == 0:
                    elapsed = time.time() - start_time
                    rate = success_count / elapsed if elapsed > 0 else 0
                    remaining = (num_patches - success_count) / rate if rate > 0 else 0
                    print(f"    Progress: {success_count}/{num_patches} "
                          f"({rate:.1f} patches/sec, ~{remaining:.0f}s remaining)")

            except Exception as e:
                print(f"    Task {idx} generated an exception: {e}")

    elapsed = time.time() - start_time
    print(f"  ✓ Completed: {success_count}/{num_patches} patches in {elapsed:.1f}s "
          f"({success_count/elapsed:.1f} patches/sec)")

    return success_count

def resize_all_patches_efficient(dbx,
                                 source_base_path='/patches',
                                 dest_base_path='/small_patches',
                                 metadata_path='/patches/metadata.json',
                                 new_size=(128, 128),
                                 max_workers=10):
    """
    Efficiently resize all patches using parallel processing
    """

    print("="*60)
    print("EFFICIENT BATCH RESIZING TO 128x128")
    print("="*60)
    print(f"Using {max_workers} parallel workers per tile")

    # Load metadata
    print("\nLoading metadata...")
    try:
        _, response = dbx.files_download(metadata_path)
        metadata = json.loads(response.content)
        print(f"✓ Metadata loaded")
        print(f"  Original size: {metadata['patch_size']}x{metadata['patch_size']}")
        print(f"  New size: {new_size[0]}x{new_size[1]}")
        print(f"  Total tiles: {len(metadata['tiles'])}")

        total_patches = sum(t['num_patches'] for t in metadata['tiles'].values())
        print(f"  Total patches to process: {total_patches}")
    except Exception as e:
        print(f"ERROR: {e}")
        return None

    # Create new metadata
    new_metadata = {
        'patch_size': new_size[0],
        'num_patches_per_tile': metadata['num_patches_per_tile'],
        'original_patch_size': metadata['patch_size'],
        'tiles': {}
    }

    overall_start = time.time()
    total_resized = 0

    # Process each tile
    for idx, (tile_id, tile_info) in enumerate(metadata['tiles'].items(), 1):
        risk = tile_info['risk_category']
        num_patches = tile_info['num_patches']

        print(f"\n{'='*60}")
        print(f"[{idx}/{len(metadata['tiles'])}] Tile: {tile_id} ({risk})")
        print(f"{'='*60}")

        # Resize this tile's patches in parallel
        resized_count = resize_tile_batch(
            dbx, tile_id, risk, num_patches,
            source_base_path, dest_base_path,
            new_size, max_workers
        )

        # Update metadata
        new_metadata['tiles'][tile_id] = {
            'risk_category': risk,
            'dimensions': tile_info['dimensions'],
            'num_patches': resized_count,
            'description': tile_info['description'],
            'coords': tile_info['coords']
        }

        total_resized += resized_count

        # Overall progress
        elapsed = time.time() - overall_start
        avg_rate = total_resized / elapsed if elapsed > 0 else 0
        print(f"\n  Overall progress: {total_resized}/{total_patches} patches "
              f"({avg_rate:.1f} patches/sec)")

    # Save metadata
    total_time = time.time() - overall_start
    print(f"\n{'='*60}")
    print("RESIZING COMPLETE")
    print(f"{'='*60}")
    print(f"Total patches: {total_resized}/{total_patches}")
    print(f"Total time: {total_time/60:.1f} minutes")
    print(f"Average speed: {total_resized/total_time:.1f} patches/sec")
    print(f"\nSaving metadata...")

    new_metadata_json = json.dumps(new_metadata, indent=2)
    new_metadata_bytes = new_metadata_json.encode('utf-8')
    dbx.files_upload(new_metadata_bytes, f'{dest_base_path}/metadata.json',
                    mode=dropbox.files.WriteMode.overwrite)

    print(f"✓ Metadata saved")

    # Memory savings
    original_size = (metadata['patch_size'] ** 2) * 3
    new_size_bytes = (new_size[0] ** 2) * 3
    reduction = ((original_size - new_size_bytes) / original_size) * 100

    print("\n" + "="*60)
    print("SUMMARY")
    print("="*60)
    print(f"Source: {source_base_path}")
    print(f"Destination: {dest_base_path}")
    print(f"Memory reduction: {reduction:.1f}% per patch")

    return new_metadata

# ============================================================================
# VERIFICATION
# ============================================================================

def verify_resized_patches(dbx, dest_base_path='/small_patches', num_samples=6):
    """
    Verify resized patches by displaying samples
    """
    import matplotlib.pyplot as plt
    import random

    print("\n" + "="*60)
    print("VERIFYING RESIZED PATCHES")
    print("="*60)

    # Load metadata
    _, response = dbx.files_download(f'{dest_base_path}/metadata.json')
    metadata = json.loads(response.content)

    print(f"Patch size in metadata: {metadata['patch_size']}x{metadata['patch_size']}")

    # Sample patches from different risk categories
    samples = []
    for risk in ['high_risk', 'medium_risk', 'low_risk']:
        # Find tiles in this category
        risk_tiles = {tid: info for tid, info in metadata['tiles'].items()
                     if info['risk_category'] == risk}
        if risk_tiles:
            tile_id = random.choice(list(risk_tiles.keys()))
            tile_info = risk_tiles[tile_id]
            patch_idx = random.randint(0, tile_info['num_patches']-1)
            samples.append((tile_id, risk, patch_idx))

    # Add more random samples if needed
    while len(samples) < num_samples:
        tile_id = random.choice(list(metadata['tiles'].keys()))
        tile_info = metadata['tiles'][tile_id]
        risk = tile_info['risk_category']
        patch_idx = random.randint(0, tile_info['num_patches']-1)
        samples.append((tile_id, risk, patch_idx))

    samples = samples[:num_samples]

    # Display samples
    fig, axes = plt.subplots(2, 3, figsize=(12, 8))
    axes = axes.flatten()

    color_map = {'high_risk': 'red', 'medium_risk': 'orange', 'low_risk': 'green'}

    for idx, (tile_id, risk, patch_idx) in enumerate(samples):
        try:
            patch_path = f"{dest_base_path}/{risk}/{tile_id}/patch_{patch_idx:04d}.png"
            _, response = dbx.files_download(patch_path)
            img = Image.open(io.BytesIO(response.content))
            img_array = np.array(img)

            axes[idx].imshow(img_array)
            axes[idx].set_title(f'{risk.replace("_", " ").title()}\n{tile_id}\n{img_array.shape[0]}x{img_array.shape[1]}')
            axes[idx].axis('off')

            # Add colored border
            for spine in axes[idx].spines.values():
                spine.set_edgecolor(color_map[risk])
                spine.set_linewidth(3)
                spine.set_visible(True)

        except Exception as e:
            print(f"ERROR loading sample: {e}")

    plt.suptitle('Resized Patches (128x128)', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

    print("✓ Verification complete")

# ============================================================================
# MAIN EXECUTION
# ============================================================================

if __name__ == "__main__":
    print("Starting EFFICIENT patch resizing...")
    print("This uses parallel processing to speed up the operation")

    # Adjust max_workers based on your needs (5-15 is usually good)
    MAX_WORKERS = 10

    # Resize all patches
    new_metadata = resize_all_patches_efficient(
        dbx=dbx,
        source_base_path='/patches',
        dest_base_path='/small_patches',
        metadata_path='/patches/metadata.json',
        new_size=(128, 128),
        max_workers=MAX_WORKERS
    )

    if new_metadata is not None:
        # Verify
        verify_resized_patches(dbx, dest_base_path='/small_patches', num_samples=6)

        print("\n" + "="*60)
        print("SUCCESS!")
        print("="*60)
        print("Resized patches available in '/small_patches'")
    else:
        print("\nProcess failed. Check errors above.")
```
